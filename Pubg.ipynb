```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("PUBG.csv")
print(df.head())
print(df.shape)
print(df.columns)
print(df.dtypes)
print(df.isnull().sum())
print(df.describe(include='all'))
```

```python
df["Map"].value_counts().plot(kind="bar")
plt.title("Number of Players by Map")
plt.xlabel("Map")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(df["Kills"], bins=10)
plt.title("Distribution of Kills")
plt.xlabel("Kills")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df["Kills"], df["Damage_Dealt"])
plt.title("Kills vs Damage Dealt")
plt.xlabel("Kills")
plt.ylabel("Damage Dealt")
plt.show()
```

```python
sns.boxplot(y=df["Damage_Dealt"])
plt.title("Box Plot of Damage Dealt")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include="number").corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df["Favorite_Weapon"].value_counts().head(5).plot(
    kind="pie", autopct="%1.1f%%"
)
plt.title("Top 5 Favorite Weapons")
plt.ylabel("")
plt.show()
```
