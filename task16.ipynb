```python
#Q.5)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("insurance.csv")
num_cols = ['age', 'bmi', 'children', 'region']

for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
#Q.6)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("insurance.csv")
cat_cols = ['sex', 'smoker', 'region']
for col in cat_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, data=df, palette='Set2')
    plt.title(f"Count Plot of {col}")
    plt.xlabel(col.capitalize())
    plt.ylabel("Count")
    plt.show()
```

```python
#Q.7)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("insurance.csv")
numerical_data = df[['age', 'bmi', 'children', 'expenses']]
correlation = numerical_data.corr()
plt.figure(figsize=(6, 5))
sns.heatmap(correlation,
            annot=True,
            cmap="coolwarm",
            fmt=".2f",
            linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Variables")
plt.show()
```

```python
#Q.8)
import numpy as np
import pandas as pd
df = pd.read_csv("insurance.csv")
average_charges = df['expenses'].mean()
print("Average Insurance expenses:", average_charges)
maximum_charges = df['expenses'].max()
print("Maximum Insurance Charges:", maximum_charges)
minimum_charges = df['expenses'].min()
print("Minimum Insurance expenses:", minimum_charges)
print("\nAverage Charges by Smoker Status:")
print(df.groupby('smoker')['expenses'].mean())
```

```python
#Q.9)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("insurance.csv")
plt.figure(figsize=(6, 4))
sns.boxplot(x='smoker', y='expenses', data=df, palette='Set2')
plt.title("Insurance Charges by Smoker")
plt.xlabel("Smoker")
plt.ylabel("expenses")
plt.show()
plt.figure(figsize=(6, 4))
sns.boxplot(x='sex', y='expenses', data=df, palette='Set3')
plt.title("Insurance expenses by Sex")
plt.xlabel("Sex")
plt.ylabel("expenses")
plt.show()
```
